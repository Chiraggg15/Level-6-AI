```python
# Task 1
def unit_converter():
    """
    Converts between length, weight, and volume units.

    The user selects the type of conversion and inputs a value.
    The function performs the conversion and returns the result.

    Returns:
        float: converted value
    """

    try:
        print("Choose conversion type:")
        print("1. Length")
        print("2. Weight")
        print("3. Volume")

        choice = input("Enter choice (1/2/3): ")

        value = float(input("Enter value to convert: "))

        if choice == "1":
            print("1. Meters to Feet")
            print("2. Feet to Meters")
            opt = input("Choose option: ")

            if opt == "1":
                result = value * 3.28084
                print("Result:", result, "ft")

            elif opt == "2":
                result = value / 3.28084
                print("Result:", result, "m")

            else:
                print("Invalid option")

        elif choice == "2":
            print("1. Kilograms to Pounds")
            print("2. Pounds to Kilograms")
            opt = input("Choose option: ")

            if opt == "1":
                result = value * 2.20462
                print("Result:", result, "lbs")

            elif opt == "2":
                result = value / 2.20462
                print("Result:", result, "kg")

            else:
                print("Invalid option")

        elif choice == "3":
            print("1. Liters to Gallons")
            print("2. Gallons to Liters")
            opt = input("Choose option: ")

            if opt == "1":
                result = value * 0.264172
                print("Result:", result, "gal")

            elif opt == "2":
                result = value / 0.264172
                print("Result:", result, "L")

            else:
                print("Invalid option")

        else:
            print("Invalid choice")

    except ValueError:
        print("Error: Please enter a valid number.")


unit_converter()
```

```python
# Task 2
def find_sum(numbers):
    """Returns sum of numbers."""
    return sum(numbers)

def find_average(numbers):
    """Returns average of numbers."""
    return sum(numbers) / len(numbers)

def find_max(numbers):
    """Returns maximum number."""
    return max(numbers)

def find_min(numbers):
    """Returns minimum number."""
    return min(numbers)


try:
    nums = list(map(float, input("Enter numbers separated by space: ").split()))

    if len(nums) == 0:
        raise ValueError("Empty list")

    print("Choose operation:")
    print("1. Sum")
    print("2. Average")
    print("3. Maximum")
    print("4. Minimum")

    op = input("Enter choice: ")

    if op == "1":
        print("Sum:", find_sum(nums))
    elif op == "2":
        print("Average:", find_average(nums))
    elif op == "3":
        print("Maximum:", find_max(nums))
    elif op == "4":
        print("Minimum:", find_min(nums))
    else:
        print("Invalid operation")

except ValueError:
    print("Error: Please enter valid numbers.")
```

```python
# 1 List Manipulation 4.2
def extract_every_other(lst):
    """Return every other element."""
    return lst[::2]

print(extract_every_other([1, 2, 3, 4, 5, 6]))
```

```python
# 2 Slice sublist
def get_sublist(lst, start, end):
    """Return sublist from start to end inclusive."""
    return lst[start:end+1]

print(get_sublist([1, 2, 3, 4, 5, 6], 2, 4))
```

```python
# 3 Reverse list
def reverse_list(lst):
    """Reverse list using slicing."""
    return lst[::-1]

print(reverse_list([1, 2, 3, 4, 5]))
```

```python
# 4 Remove first and last
def remove_first_last(lst):
    """Remove first and last element."""
    return lst[1:-1]

print(remove_first_last([1, 2, 3, 4, 5]))
```

```python
# 5 first n elements
def get_first_n(lst, n):
    """Return first n elements."""
    return lst[:n]

print(get_first_n([1, 2, 3, 4, 5], 3))
```

```python
# 6 last n elements
def get_last_n(lst, n):
    """Return last n elements."""
    return lst[-n:]

print(get_last_n([1, 2, 3, 4, 5], 2))
```

```python
# 7 reverse skip
def reverse_skip(lst):
    """Return every second element from second last backwards."""
    return lst[-2::-2]

print(reverse_skip([1, 2, 3, 4, 5, 6]))
```

```python
# Nested list exercise 4.3
# 1 flatten nested list
def flatten(lst):
    """Flatten nested list."""
    result = []
    for sub in lst:
        result.extend(sub)
    return result

print(flatten([[1, 2], [3, 4], [5]]))
```

```python
# 2 Access nested element
def access_nested_element(lst, indices):
    """Access element using index list."""
    element = lst
    for i in indices:
        element = element[i]
    return element

lst = [[1,2,3],[4,5,6],[7,8,9]]
print(access_nested_element(lst, [1,2]))
```

```python
# 3 Sum nested
def sum_nested(lst):
    """Sum elements in nested list."""
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item
    return total

print(sum_nested([[1,2],[3,[4,5]],6]))
```

```python
# 4 remove element
def remove_element(lst, elem):
    """Remove element from nested list."""
    result = []
    for item in lst:
        if isinstance(item, list):
            result.append(remove_element(item, elem))
        elif item != elem:
            result.append(item)
    return result

print(remove_element([[1,2],[3,2],[4,5]], 2))
```

```python
# 5 find max
def find_max(lst):
    """Find maximum in nested list."""
    max_val = float("-inf")
    for item in lst:
        if isinstance(item, list):
            max_val = max(max_val, find_max(item))
        else:
            max_val = max(max_val, item)
    return max_val

print(find_max([[1,2],[3,[4,5]],6]))
```

```python
# 6 count occurance
def count_occurrences(lst, elem):
    """Count element in nested list."""
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        elif item == elem:
            count += 1
    return count

print(count_occurrences([[1,2],[2,3],[2,4]], 2))
```

```python
# 7 deep flatten
def deep_flatten(lst):
    """Flatten list regardless of depth."""
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(deep_flatten(item))
        else:
            result.append(item)
    return result

print(deep_flatten([[[1,2],[3,4]],[[5,6],[7,8]]]))
```

```python
# 8 nested average
def average_nested(lst):
    """Return average of nested list."""
    flat = deep_flatten(lst)
    return sum(flat) / len(flat)

print(average_nested([[1,2],[3,4],[5,6]]))
```
